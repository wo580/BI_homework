```python
import pandas as pd
from sklearn.preprocessing import *
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#Step1，数据加载
df = pd.read_csv('./voice.csv')
pd.set_option('display.max_columns',1000)
print(df)
```

```python
print(df.isnull().sum())#缺失值
print(df.shape)#大小
print('样本个数：{}'.format(df.shape[0]))
print('男个数：{}'.format(df[df.label=='male'].shape[0]))
print('女个数：{}'.format(df[df.label=='female'].shape[0]))
```

```python
# Step2，数据预处理
# 分离特征X 和Target y
#:-1不取最后一列lable
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
# 使用标签编码，male -> 1, female -> 0
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
print(y)
print(X)
```

```python
# 将特征X矩阵进行规范化
# #标准差标准化，处理后的数据符合标准正态分布
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#Step3，数据集切分，train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Step4，模型训练
svc=SVC()
svc.fit(X_train,y_train)
```

```python
#Step5，模型预测
y_pred = svc.predict(X_test)
print('SVM 预测结果:',y_pred)
print('SVM 预测准确率:',accuracy_score(y_test,y_pred))
print('SVM 预测准确率:',svc.score(X_test, y_pred))
```

```python
#linear SVC预测
scv=SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print('linear SVM 预测结果:',y_pred)
print('linear SVM 预测准确率:',accuracy_score(y_test,y_pred))
print('linear SVM 预测准确率:',svc.score(X_test, y_pred))
```

```python
#使用xgboost预测
import xgboost as xgb
param = {'boosting_type':'gbdt',
                         'objective' : 'binary:logistic', #任务目标
                         'eval_metric' : 'auc', #评估指标
                         'eta' : 0.01, #学习率
                         'max_depth' : 15, #树最大深度
                         'colsample_bytree':0.8, #设置在每次迭代中使用特征的比例
                         'subsample': 0.9, #样本采样比例
                         'subsample_freq': 8, #bagging的次数
                         'alpha': 0.6, #L1正则
                         'lambda': 0, #L2正则
        }
```

```python
train_data = xgb.DMatrix(X_train,y_train)
test_data = xgb.DMatrix(X_test,y_test)

model = xgb.train(param, train_data, evals=[(train_data,'train'),(test_data,'valid')],num_boost_round=3000,early_stopping_rounds=25,verbose_eval=25)
```

```python
y_pred = model.predict(test_data)
y_pred = [1 if x>=0.5 else 0 for x in y_pred]#2分法
print('XGBoost预测结果：',y_pred)
print('XGBoost预测准确率：',accuracy_score(y_test,y_pred))
```
