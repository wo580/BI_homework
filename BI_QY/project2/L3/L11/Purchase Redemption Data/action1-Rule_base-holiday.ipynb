预测下个月每一天的情况：
如果想预测下个月每天的流量情况，可以基于每月的规律（1-30号的平均流量）* 周期因子
Step1，计算周期因子（weekday）
Step2，计算每日（1号-30号）均值，即1号的平均流量，2号的平均流量 …
Step3，统计星期几（weekday）在每日（day）出现的频次
Step4，基于周期因子获得加权均值，得到每日的base（去掉周期因子的影响）
Step5，根据每日的base和周期因子进行预测

# 数据探索

```python
#数据加载,时间序列
import pandas as pd
#parse_dates=['report_date']转换成日期类型
data = pd.read_csv('./user_balance_table.csv') 
data
```

```python
#给时间添加时间维度
def add_timestamp(data):
    #时间格式转换
    data['report_date'] = pd.to_datetime(data['report_date'],format="%Y%m%d")
    #添加时间维度
    data['day'] = data['report_date'].dt.day
    data['month'] = data['report_date'].dt.month
    data['year'] = data['report_date'].dt.year
    data['week'] = data['report_date'].dt.week#一年中的第几个星期
    data['weekday'] = data['report_date'].dt.weekday#0代表周一,6周日
    return data

data=add_timestamp(data)
data
```

```python
def get_total_balance(data,begin):
    #在copy数据上进行操作
    df_temp = data.copy()
    #按照data进行聚合
    df_temp = df_temp.groupby(['report_date'])['total_purchase_amt','total_redeem_amt'].sum()
    #print(df_temp)
    
    #还原data字段,重新索引
    df_temp.reset_index(inplace=True)
    df_temp = df_temp[(df_temp['report_date']>=begin)]
    return df_temp
#筛选从2014-04-01之后的数据作为平稳数据
total_balance = get_total_balance(data,'2014-03-01')
total_balance
```

```python
import datetime
import numpy as np
#生成测试数据
def generate_test_data(data):
    #在copy数据上进行操作
    total_balance = data.copy()
    #生成2014-09-01到2014-09-30的数据
    start =datetime.datetime(2014,9,1)
    end = datetime.datetime(2014,10,1)
    testdata=[]
    while start!=end:
        #添加3个字段,date,total_purchase_amt,total_redeem_amt
        temp = [start,np.nan,np.nan]
        testdata.append(temp)
        #日期+1
        start += datetime.timedelta(days=1)
        
    #封装testdata
    testdata = pd.DataFrame(testdata)
    testdata.columns = total_balance.columns
    #将testdata合并到total_balance中
    total_balance = pd.concat([total_balance,testdata],axis=0)
    return total_balance
total_balance = generate_test_data(total_balance)    
total_balance
```

```python
#对total_balance 添加时间维度
total_balance  = add_timestamp(total_balance)
total_balance 
```

```python
#修正日期假期
import datetime
from chinese_calendar import is_workday, is_holiday
total_balance['is_holiday'] =total_balance['report_date'].apply(lambda x: is_holiday(x))
#将True和False映射为1和0
total_balance['is_holiday'] =total_balance['is_holiday'].replace({True:1,False:0})
total_balance
```

```python
#检查异常日期, 工作日为假日延长放假日, 放假日为假期加班日
#re_weekday=[]
#for index,(weekday, is_holiday) in enumerate(zip(total_balance['weekday'].values, total_balance['is_holiday'].values)):
#    r = total_balance['weekday'].values[index]
#    #如果不是周六日, 但是是holiday=>设为周日
#    if weekday not in[5,6] and is_holiday ==1:
#        r=6
#    #如果是周六日, 但是上班=>设为周一
#    elif  weekday in[5,6] and is_holiday ==0:
#        r=0
#    re_weekday.append(r)
#total_balance['weekday'] = re_weekday
#total_balance
```

```python
#检查异常日期, 工作日为假日延长放假日, 放假日为假期加班日
re_weekday=[]
for index,(weekday, is_holiday) in enumerate(zip(total_balance['weekday'].values, total_balance['is_holiday'].values)):
    r = total_balance['weekday'].values[index]
    #如果不是周六日, 但是是holiday=>设为周日
    if weekday not in[5,6] and is_holiday ==1:
        r=6
    #如果是周六日, 但是上班=>设为周一
    elif  weekday in[5,6] and is_holiday ==0:
        r=0
    re_weekday.append(r)
total_balance['weekday'] = re_weekday
total_balance
```

```python
temp = total_balance.copy()#备份集
```

```python
#复制一份作为操作集
total_balance  = temp.copy()
total_balance
```

```python
#total_balance = total_balance[['report_date','total_purchase_amt','total_redeem_amt']]
weekday_weight = total_balance[['weekday','total_purchase_amt','total_redeem_amt']].groupby('weekday',as_index=False)
weekday_weight = weekday_weight.mean()
weekday_weight
```

```python
weekday_weight = weekday_weight.rename(columns={'total_purchase_amt':'purchase_weekday','total_redeem_amt':'redeem_weekday'})
weekday_weight
```

```python
#计算周期因子
weekday_weight['purchase_weekday'] /= np.mean(total_balance['total_purchase_amt'])
weekday_weight['redeem_weekday'] /= np.mean(total_balance['total_redeem_amt'])
weekday_weight
```

```python
#合并
total_balance = pd.merge(total_balance,weekday_weight,on='weekday',how='left')
total_balance
```

```python
#计算日期因子#统计周一~周日,在1-31号中出现的次数
weekday_count = total_balance[['report_date','day','weekday']].groupby(['day','weekday'],as_index=False).count()
```

```python
weekday_count = pd.merge(weekday_count,weekday_weight,on='weekday')
weekday_count
```

```python
len(np.unique(total_balance['month']))
```

```python
#日期计算 = 周期因子 * (周一到周日在1-31号中出现的频次)/一共多少个月
weekday_count['purchase_weekday'] = weekday_count['purchase_weekday']*weekday_count['report_date']/len(np.unique(total_balance['month']))
```

```python
weekday_count['redeem_weekday'] = weekday_count['redeem_weekday']*weekday_count['report_date']/len(np.unique(total_balance['month']))
```

```python
#日期因子
day_rate = weekday_count.drop(['weekday','report_date'],axis=1).groupby('day',as_index=False).sum()
day_rate
```

```python
#按照日期day计算均值,即1号的均值,2号的均值...
day_mean = total_balance[['day','total_purchase_amt','total_redeem_amt']].groupby('day',as_index=False).mean()
day_mean
```

```python
day_pred = pd.merge(day_mean,day_rate,on='day',how='left')
day_pred
```

```python
#去掉周期因子影响,之后的日期因子
day_pred['total_purchase_amt']/=day_pred['purchase_weekday']
day_pred['total_redeem_amt']/=day_pred['redeem_weekday']
day_pred
```

```python
for index,row in day_pred.iterrows():
    if row['day']==31:
        break
    #添加日期字段report_date
    day_pred.loc[index,'report_date'] = pd.to_datetime('2014/09/'+str(int(row['day'])))
day_pred
```

```python
#利用周期因子,进行base*周期因子=预测结果
day_pred['weekday'] = day_pred['report_date'].dt.weekday
day_pred
```

```python
day_pred=day_pred[['day','total_purchase_amt','total_redeem_amt','report_date','weekday']]
day_pred
```

```python
day_pred = pd.merge(day_pred,weekday_weight,on='weekday')
day_pred
```

```python
day_pred['total_purchase_amt']*= day_pred['purchase_weekday']
day_pred['total_redeem_amt']*= day_pred['redeem_weekday']
day_pred
```

```python
#排序
day_pred = day_pred[['report_date','total_purchase_amt','total_redeem_amt']].sort_values('report_date')
day_pred
```

```python
day_pred['report_date'] = day_pred['report_date'].apply(lambda x:str(x).replace('-','')[0:8])#日期格式去掉-
```

```python
day_pred.to_csv('rule_base_chinese.csv',header=None,index=False)#提交查看成绩
```
